```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dates=pd.date_range(start='2023-01-01',end='2023-04-10',freq='d')
```

```python
values=np.random.randn(100)
```

```python
data=pd.DataFrame(values,index=dates,columns=['Values'])
```

```python
data.head()
```

```python
data.plot(title='Time vs Values',color='gold')
plt.xlabel("Time")
plt.ylabel("Values")
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
fig,axes=plt.subplots(2,1,figsize=(8,6))
plot_acf(data['Values'],lags=30,ax=axes[0],title='Autocorrelation Function(ACF)')
plot_pacf(data['Values'],lags=30,ax=axes[1],title='Partial Autocorrelation Function(PACF)')
plt.tight_layout()
plt.show()
```
