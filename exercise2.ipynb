```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(40)
```

```python
values=np.random.randn(100)
```

```python
values
```

```python
dates=pd.date_range(start='2023-01-01',end='2023-04-10',freq='D')
```

```python
dates
```

```python
data=pd.DataFrame(values,index=dates,columns=['value'])
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('Date')
plt.ylabel('Value')
plt.plot(data['value'])
plt.title('Dates Vs Values')
plt.show()
```

```python
data['year']=data.index.year
```

```python
data['lag1']=data['value'].shift(1)
data['lag2']=data['value'].shift(2)
```

```python
data['rollingmean']=data['value'].rolling(5).mean()
data['rollingstd']=data['value'].rolling(5).std()
data['rollingmin']=data['value'].rolling(5).min()
data['rollingmax']=data['value'].rolling(5).max()
```

```python
data.tail()
```

```python
data['expandingmean']=data['value'].expanding().mean()
data['expandingstd']=data['value'].expanding().mean()
data['exapndingmin']=data['value'].expanding().min()
data['exapndingmax']=data['value'].expanding().max()
```
